```python
import pandas as pd

# Example DataFrame
data = {'sentence': ['This is a sentence with word1 and word3 and word1.',
                     'Another sentence with word2 and word1.',
                     'No matching words here.']}
df = pd.DataFrame(data)

# Example word values dictionary
word_values = {'word1': 10, 'word2': 5, 'word3': 8}

# Threshold values and corresponding labels
thresholds = {50: 'High', 20: 'Medium', 10: 'Low', 5: 'Very Low'}

# Dictionary to store results
sentence_word_counts = {}

# Iterate through each sentence in the DataFrame
for index, row in df.iterrows():
    sentence = row['sentence']
    word_count = {}

    # Check if each word in the dictionary is present in the sentence
    for word, value in word_values.items():
        # Count the occurrences of each word in the sentence
        count = sentence.count(word)
        if count > 0:
            word_count[word] = count

    # Sum the values corresponding to the words in the sentence
    total_value = sum(word_values[word] * count for word, count in word_count.items())

    # Determine the label based on the total value and thresholds
    label = None
    for threshold, threshold_label in thresholds.items():
        if total_value >= threshold:
            label = threshold_label
            break

    # Store the results in the dictionary
    sentence_word_counts[index] = {'sentence': sentence, 'word_count': word_count, 'total_value': total_value, 'label': label}

# Create a DataFrame from the dictionary
result_df = pd.DataFrame.from_dict(sentence_word_counts, orient='index')

# Display the results
result_df
```
